```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data = pd.read_csv('/Users/b1017579/Documents/PhD/Projects/10. ELECSIM/elecsim/data/processed/multi_day_data/8_medoids.csv')
data = pd.read_csv('/Users/b1017579/Documents/PhD/Projects/10. ELECSIM/elecsim/data/processed/multi_day_data/4_medoids.csv')
data
```

```python
# np.round(np.unique(data[data.data_type=="solar"].counts))
```

```python
data = data.rename(columns={"data_type":"Data Type"})
data['Data Type'] = data['Data Type'].map({"load":"Load", 'solar':"Solar","offshore":"Offshore",'onshore':"Onshore"})
data
```

```python
sns.set(font_scale=1.5)
sns.set_style("whitegrid")
g=sns.lineplot(data=data[data['Data Type']=='Load'], x='index', y='capacity_factor', hue='Data Type')
g.set(xlabel='Hours', ylabel='Capacity Factor')
lgd = plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
g.set_xticks(list(range(0,192,24)))
g.set(ylim=(0))
plt.savefig('/Users/b1017579/Documents/PhD/Projects/14-temporal-granularity/temporal_granularity/reports/figures/clusters_results_load.pdf', dpi=500, bbox_extra_artists=(lgd,), bbox_inches='tight')
```

```python
sns.set(font_scale=1.5)
sns.set_style("whitegrid")
g=sns.lineplot(data=data[data['Data Type']!='Load'], x='index', y='capacity_factor', hue='Data Type')
g.set(xlabel='Hours', ylabel='Capacity Factor')
lgd = plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
g.set_xticks(list(range(0,192,24)))
plt.savefig('/Users/b1017579/Documents/PhD/Projects/14-temporal-granularity/temporal_granularity/reports/figures/clusters_results_resources.pdf', dpi=500, bbox_extra_artists=(lgd,), bbox_inches='tight')
```
